```python
# CÉLULA 7 — Abrindo o dicionário de variáveis

caminho_dicionario = os.path.join(pasta_doc, "Dicionário 1991.xls")

dicionario = pd.read_excel(caminho_dicionario)
dicionario.head(20)
```

```python
# CÉLULA 8 — Filtrando variáveis de interesse

variaveis_interesse = [
    "RELIGIAO", "RACACOR", "EDGRAU", "RTOTALPV", "IDADEANO", "SEXO",
    "SITSET", "POSOCUP", "OCUPAGRP"
]

for var in variaveis_interesse:
    print(f"\n{'='*40}")
    print(f"VARIÁVEL: {var}")
    print('='*40)
    
    linha_var = dicionario[dicionario["Variável"] == var]
    if not linha_var.empty:
        idx = linha_var.index[0]
        print(dicionario.iloc[idx])
        
        i = idx + 1
        while i < len(dicionario):
            proxima_variavel = dicionario.iloc[i]["Variável"]
            if pd.notna(proxima_variavel):
                break
            print(dicionario.iloc[i][["Tipo", "Alias"]].values)
            i += 1
    else:
        print("Variável não encontrada!")
```

```python
# CÉLULA 9 — Função de classificação de religião (1991)
# O Censo 1991 usa códigos de 2 dígitos (dezena = grupo religioso)
# Nota: o layout DBF de 1991 não possui dicionário oficial público para
# esta codificação; o mapeamento por faixas foi validado contra as
# proporções históricas conhecidas (espiritismo ~1,1% em 1991)

def classificar_religiao_1991(codigo):
    if codigo == 0:
        return "Sem religião"
    elif 11 <= codigo <= 14:
        return "Católica"
    elif 21 <= codigo <= 29:
        return "Evangélica Tradicional"
    elif 31 <= codigo <= 39:
        return "Evangélica Pentecostal"
    elif 41 <= codigo <= 49:
        return "Outras Cristãs"
    elif 51 <= codigo <= 69:
        return "Mediúnica/Espírita"
    elif 71 <= codigo <= 89:
        return "Outras"
    elif codigo == 99:
        return "Sem declaração"
    else:
        return "Não classificado"

print("Função 1991 pronta!")
```

```python
# CÉLULA 10 — Função para ler uma amostra de um DBF

colunas = ["RELIGIAO", "RACACOR", "EDGRAU", "EDANOEST", "RTOTALPV", "IDADEANO", "SEXO",
           "SITSET", "POSOCUP", "OCUPAGRP"]

def ler_amostra_dbf(caminho_dbf, n_amostra=5000):
    tabela = DBF(caminho_dbf, encoding="latin1")
    registros = list(islice(tabela, n_amostra))
    df = pd.DataFrame(registros)
    df = df[colunas]
    df["grupo_religioso"] = df["RELIGIAO"].apply(classificar_religiao_1991)
    return df

print("Função pronta!")
```

```python
# CÉLULA 11 — Lendo amostra de todas as regiões de 1991

regioes_1991 = os.listdir(pasta_dados)
print("Regiões encontradas:", regioes_1991)

lista_dfs = []

for regiao in regioes_1991:
    pasta_regiao = os.path.join(pasta_dados, regiao)
    arquivos_dbf = [f for f in os.listdir(pasta_regiao) if f.endswith(".DBF")]
    
    for arquivo in arquivos_dbf:
        caminho = os.path.join(pasta_regiao, arquivo)
        print(f"Lendo {regiao}/{arquivo}...")
        df_temp = ler_amostra_dbf(caminho)
        df_temp["regiao"] = regiao
        df_temp["arquivo_origem"] = arquivo
        lista_dfs.append(df_temp)

df_1991 = pd.concat(lista_dfs, ignore_index=True)
df_1991["ano"] = 1991

print(f"\nTotal de registros (1991): {len(df_1991)}")
df_1991.head()
```

```python
# CÉLULA 12 — Resumo de 1991

print(f"Total de registros: {len(df_1991)}")
print(f"\nDistribuição por grupo religioso:")
print(df_1991["grupo_religioso"].value_counts())
print(f"\nDistribuição por região:")
print(df_1991["regiao"].value_counts())
```

```python
# CÉLULA 13 — Caminhos de 2000

pasta_2000 = os.path.join(DATA_DIR, "2000")

print("Caminhos de 2000 definidos!")
```

```python
# CÉLULA 14 — Função de classificação de religião (2000 e 2010)
# Os censos 2000 e 2010 usam códigos de 3 dígitos com a mesma estrutura
# Fonte: arquivo oficial "Religião 2010.txt" (Anexos Auxiliares do IBGE)

def classificar_religiao_3dig(codigo):
    if pd.isna(codigo):
        return None
    codigo = int(codigo)
    
    if codigo <= 2:
        return "Sem religião"
    elif 100 <= codigo <= 199:
        return "Católica"
    elif 200 <= codigo <= 300:
        return "Evangélica Tradicional"
    elif 310 <= codigo <= 489:
        return "Evangélica Pentecostal"
    elif 490 <= codigo <= 499:
        return "Evangélica não determinada"
    elif 510 <= codigo <= 539:
        return "Outras Cristãs"
    elif 590 <= codigo <= 649:
        return "Mediúnica/Espírita"
    elif 710 <= codigo <= 719:
        return "Judaica"
    elif 740 <= codigo <= 799:
        return "Oriental"
    elif 810 <= codigo <= 839:
        return "Outras"
    elif codigo == 850:
        return "Outras Cristãs"
    elif codigo >= 890:
        return "Sem declaração"
    else:
        return "Não classificado"

print("Função de classificação (3 dígitos) pronta!")
```

```python
# CÉLULA 15 — Função de leitura 2000
# Posições extraídas do layout oficial (LE PESSOAS.sas)

colunas_2000 = {
    "uf":          (1, 2),
    "sitset":      (65, 1),
    "sexo":        (69, 1),
    "idade":       (79, 3),
    "racacor":     (87, 1),
    "religiao":    (89, 3),
    "anos_estudo": (168, 2),
    "renda_total": (309, 6),
    "renda_sm":    (315, 6),
}

def ler_pessoas_2000(caminho_txt, n_amostra=5000):
    registros = []
    arquivo = open(caminho_txt, "r", encoding="latin1")
    
    contador = 0
    for linha in arquivo:
        if contador >= n_amostra:
            break
        registro = {}
        for nome, (pos, tam) in colunas_2000.items():
            valor = linha[pos-1 : pos-1+tam].strip()
            registro[nome] = valor
        registros.append(registro)
        contador = contador + 1
    
    arquivo.close()
    
    df = pd.DataFrame(registros)
    
    for col in ["idade", "anos_estudo", "renda_total", "renda_sm", "religiao"]:
        df[col] = pd.to_numeric(df[col], errors="coerce")
    
    df["renda_sm"] = df["renda_sm"] / 100
    df["grupo_religioso"] = df["religiao"].apply(classificar_religiao_3dig)
    
    return df

print("Função de leitura 2000 pronta!")
```

```python
# CÉLULA 16 — Leitura de todos os estados 2000

import glob

arquivos_2000 = glob.glob(os.path.join(pasta_2000, "**", "PES*.txt"), recursive=True)
arquivos_2000 = arquivos_2000 + glob.glob(os.path.join(pasta_2000, "**", "Pes*.txt"), recursive=True)
arquivos_2000 = list(set(arquivos_2000))  # remove duplicados

print(f"Arquivos encontrados: {len(arquivos_2000)}")

lista_2000 = []

for caminho in arquivos_2000:
    nome_pasta = os.path.basename(os.path.dirname(caminho))
    print(f"Lendo {nome_pasta}...")
    df_temp = ler_pessoas_2000(caminho)
    df_temp["uf_sigla"] = nome_pasta
    lista_2000.append(df_temp)

df_2000 = pd.concat(lista_2000, ignore_index=True)
df_2000["ano"] = 2000

print(f"\nTotal: {len(df_2000)} registros")
print(df_2000["grupo_religioso"].value_counts())
```

```python
# CÉLULA 17 — Caminhos e função de leitura 2010
# Posições extraídas do layout oficial (Layout_microdados_Amostra.xls, aba PESS)

pasta_2010 = os.path.join(DATA_DIR, "2010")

colunas_2010 = {
    "uf":          (1, 2),
    "sitset":      (53, 1),
    "sexo":        (58, 1),
    "idade":       (62, 3),
    "racacor":     (68, 1),
    "escolaridade":(158, 1),
    "renda_total": (263, 7),
    "renda_sm":    (270, 9),
    "religiao":    (396, 3),
}

def ler_pessoas_2010(caminho_txt, n_amostra=5000):
    registros = []
    arquivo = open(caminho_txt, "r", encoding="latin1")

    contador = 0
    for linha in arquivo:
        if contador >= n_amostra:
            break
        registro = {}
        for nome, (pos, tam) in colunas_2010.items():
            valor = linha[pos-1 : pos-1+tam].strip()
            registro[nome] = valor
        registros.append(registro)
        contador = contador + 1

    arquivo.close()

    df = pd.DataFrame(registros)

    for col in ["idade", "escolaridade", "renda_total", "renda_sm", "religiao"]:
        df[col] = pd.to_numeric(df[col], errors="coerce")

    df["renda_sm"] = df["renda_sm"] / 100000

    df["grupo_religioso"] = df["religiao"].apply(classificar_religiao_3dig)

    return df

print("Função de leitura 2010 pronta!")
```

```python
# CÉLULA 18 — Leitura de todos os estados 2010

arquivos_2010 = glob.glob(os.path.join(pasta_2010, "**", "Amostra_Pessoas_*.txt"), recursive=True)

print(f"Arquivos encontrados: {len(arquivos_2010)}")

lista_2010 = []

for caminho in arquivos_2010:
    nome_pasta = os.path.basename(os.path.dirname(caminho))
    print(f"Lendo {nome_pasta}...")
    df_temp = ler_pessoas_2010(caminho)
    df_temp["uf_sigla"] = nome_pasta
    lista_2010.append(df_temp)

df_2010 = pd.concat(lista_2010, ignore_index=True)
df_2010["ano"] = 2010

print(f"\nTotal: {len(df_2010)} registros")
print(df_2010["grupo_religioso"].value_counts())
```

```python
# CÉLULA 19 — Resumo dos três censos

print(f"1991: {len(df_1991)} registros")
print(f"2000: {len(df_2000)} registros")
print(f"2010: {len(df_2010)} registros")
print(f"\nTotal geral: {len(df_1991) + len(df_2000) + len(df_2010)} registros")
```

```python
# CÉLULA 20 — Harmonizando 1991

df_1991_h = pd.DataFrame()
df_1991_h["ano"] = df_1991["ano"]
df_1991_h["uf"] = df_1991["arquivo_origem"].str[9:11]  # extrai código UF do nome do arquivo
df_1991_h["sexo"] = pd.to_numeric(df_1991["SEXO"], errors="coerce")
df_1991_h["idade"] = pd.to_numeric(df_1991["IDADEANO"], errors="coerce")
df_1991_h["raca"] = pd.to_numeric(df_1991["RACACOR"], errors="coerce")
df_1991_h["escolaridade_original"] = pd.to_numeric(df_1991["EDANOEST"], errors="coerce")
df_1991_h["renda"] = pd.to_numeric(df_1991["RTOTALPV"], errors="coerce")
df_1991_h["urbano_rural"] = pd.to_numeric(df_1991["SITSET"], errors="coerce")
df_1991_h["grupo_religioso"] = df_1991["grupo_religioso"]

# Trata o código de "ignorado" da renda (99999998)
df_1991_h.loc[df_1991_h["renda"] >= 99999990, "renda"] = None

print(f"1991 harmonizado: {len(df_1991_h)} registros")
df_1991_h.head()
```

```python
# CÉLULA 21 — Harmonizando 2000 e 2010

# 2000
df_2000_h = pd.DataFrame()
df_2000_h["ano"] = df_2000["ano"]
df_2000_h["uf"] = df_2000["uf"]
df_2000_h["sexo"] = pd.to_numeric(df_2000["sexo"], errors="coerce")
df_2000_h["idade"] = df_2000["idade"]
df_2000_h["raca"] = pd.to_numeric(df_2000["racacor"], errors="coerce")
df_2000_h["escolaridade_original"] = df_2000["anos_estudo"]
df_2000_h["renda"] = df_2000["renda_total"]
df_2000_h["renda_sm"] = df_2000["renda_sm"]
df_2000_h["urbano_rural"] = pd.to_numeric(df_2000["sitset"], errors="coerce")
df_2000_h["grupo_religioso"] = df_2000["grupo_religioso"]

# 2010
df_2010_h = pd.DataFrame()
df_2010_h["ano"] = df_2010["ano"]
df_2010_h["uf"] = df_2010["uf"]
df_2010_h["sexo"] = pd.to_numeric(df_2010["sexo"], errors="coerce")
df_2010_h["idade"] = df_2010["idade"]
df_2010_h["raca"] = pd.to_numeric(df_2010["racacor"], errors="coerce")
df_2010_h["escolaridade_original"] = df_2010["escolaridade"]
df_2010_h["renda"] = df_2010["renda_total"]
df_2010_h["renda_sm"] = df_2010["renda_sm"]
df_2010_h["urbano_rural"] = pd.to_numeric(df_2010["sitset"], errors="coerce")
df_2010_h["grupo_religioso"] = df_2010["grupo_religioso"]

print(f"2000 harmonizado: {len(df_2000_h)} registros")
print(f"2010 harmonizado: {len(df_2010_h)} registros")
```

```python
# CÉLULA 22 — Renda em salários mínimos (1991) e unificação final

# Salário mínimo em setembro/1991 (data de referência do censo): Cr$ 42.000
df_1991_h["renda_sm"] = df_1991_h["renda"] / 42000

# Junta os três censos
df_final = pd.concat([df_1991_h, df_2000_h, df_2010_h], ignore_index=True)

print(f"Dataframe final: {len(df_final)} registros")
print(f"\nRegistros por ano:")
print(df_final["ano"].value_counts().sort_index())
print(f"\nColunas: {df_final.columns.tolist()}")
```

```python
# CÉLULA 23 — Salvando o dataframe final

pasta_processados = os.path.join(BASE_DIR, "data", "processed")
os.makedirs(pasta_processados, exist_ok=True)

caminho_final = os.path.join(pasta_processados, "censo_religiao_renda.csv")
df_final.to_csv(caminho_final, index=False)

print(f"Salvo em: {caminho_final}")
print(f"Tamanho: {os.path.getsize(caminho_final) / 1024 / 1024:.1f} MB")
```
