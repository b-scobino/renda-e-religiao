## 1. Introdução

Pertencer a uma religião pode ser benéfico para a renda?

A pergunta, que à primeira vista parece pertencer apenas à fé, é na verdade uma antiga
dúvida debatida pela sociologia e economia. Max Weber (1905), em *A Ética Protestante
e o Espírito do Capitalismo*, argumentou que a ética calvinista: disciplina, poupança
e trabalho como vocação, teria criado as condições culturais para a acumulação
capitalista. Mais de um século depois, a literatura segue dividida: Putnam (1993)
destaca o papel das comunidades religiosas como geradoras de capital social, redes de
confiança que se convertem em oportunidade econômica; Iannaccone (1998) inaugura a
"economia da religião", tratando a afiliação religiosa como escolha racional com
custos e retornos mensuráveis.

No Brasil, a questão ganha uma especificidade própria. O país viveu, entre 1991 e 2010,
a mais rápida transição religiosa de sua história: a população católica caiu de cerca
de 83% para 65%, enquanto os evangélicos, principalmente pentecostais e neopentecostais,
mais que dobraram de participação (Mariano, 2004).

Parte expressiva desse crescimento ocorreu devido à Teologia da Prosperidade, doutrina
segundo a qual a fé se manifesta também em prosperidade material. Trata-se, em essência,
de uma hipótese empírica declarada em cultos por seus pastores: a adesão religiosa, o
seguimento de sua hermenêutica e as ofertas financeiras constantes gerariam riqueza
para o fiel. A fé, se forte o bastante, seria recompensada com riquezas.

Este trabalho pergunta se os dados confirmam o disposto por essa doutrina.

A análise considera ainda uma distinção da análise weberiana frequentemente ignorada no debate: o
protestantismo que Weber estudou não é o pentecostalismo brasileiro contemporâneo. As
denominações tradicionais (luteranas, presbiterianas, batistas, metodistas), muitas
delas com longa tradição de instituições educacionais próprias, podem guardar relação
distinta com a renda em comparação às igrejas pentecostais, cuja expansão se deu
majoritariamente na base da pirâmide social.

A partir dos microdados dos Censos Demográficos de 1991, 2000 e 2010, este trabalho
investiga a associação entre afiliação religiosa, escolaridade e renda no Brasil,
utilizando coortes sintéticas e modelos de regressão com controles demográficos.
Duas hipóteses orientam a análise:

> **HIPÓTESE 1:** Controlando por idade, sexo, raça, escolaridade e região, indivíduos
> evangélicos pentecostais apresentam renda inferior à de outros grupos religiosos
> relevantes no país.

> **HIPÓTESE 2:** A associação entre afiliação religiosa e renda é majoritariamente
> explicada pelo perfil socioeconômico de quem adere a cada religião, e não
> por efeito direto da religião sobre a renda.

Em outras palavras: se a desvantagem de renda dos pentecostais diminuir substancialmente
quando comparados indivíduos de mesma raça, escolaridade, idade, sexo e região, a
diferença bruta reflete a composição do grupo, não é um efeito da fé professada.

Adicionalmente, explora-se uma hipótese secundária: a de que protestantes históricos se
aproximam da tese weberiana original, apresentando indicadores socioeconômicos superiores
aos de pentecostais, o que deslocaria o debate da pergunta "a fé enriquece?" para uma
formulação mais precisa: *qual* fé, em *qual* mecanismo e para *quem*.

Antecipando o resultado central: a religião, por si só, altera pouco o salário de pessoas
demograficamente equivalentes. Seu efeito relevante é indireto, está na chance de chegar
ao diploma universitário, que é o que de fato multiplica a renda no Brasil.

## 2. Metodologia

### 2.1 Método e justificativa

A investigação utiliza os microdados da amostra dos Censos Demográficos do IBGE de
1991, 2000 e 2010, a única fonte brasileira que combina, no nível do indivíduo,
religião declarada, renda, escolaridade, características demográficas e cobertura
nacional ao longo de duas décadas.

Os censos, contudo, não constituem um painel: cada edição entrevista uma amostra nova
e anônima, impossibilitando acompanhar o mesmo indivíduo ao longo do tempo.

Para contornar essa limitação, adota-se a técnica de **coortes sintéticas**
(pseudo-painel): acompanha-se o *grupo* demográfico, não a pessoa. A geração nascida
entre 1956 e 1966, por exemplo, tinha 25–35 anos em 1991, 34–44 em 2000 e 44–54 em
2010. Filtrando essas janelas etárias em cada censo, observa-se a trajetória econômica
média do mesmo grupo populacional. Entretanto, a interpretação exige um cuidado: a
composição religiosa interna da coorte pode mudar entre censos devido a conversões
religiosas, de modo que variações de renda por grupo religioso combinam
efeito-trajetória e efeito-composição. Essa migração agregada é medida e observada nos resultados.

Duas coortes adicionais exploram o mecanismo educacional: indivíduos que tinham 5 a 15
anos em 1991, cuja religião declarada reflete, na prática, a do domicílio de origem,
chamado aqui de "religião de berço", são observados na faixa de 24 a 34 anos em 2010,
quando o ciclo educacional típico está concluído. Isso permite associar a criação
religiosa ao destino educacional e à renda no início da vida adulta.

A renda é padronizada em **salários mínimos** de cada época, neutralizando as trocas
de moeda ocorridas e a inflação do período. Privilegia-se a **mediana** como medida
central, dada a forte assimetria da distribuição de renda brasileira; médias são
reportadas como complemento.

### 2.2 Estratégia estatística

A análise progride em três níveis de exigência:

**(a) Estatística descritiva comparativa.** Renda mediana e média por grupo religioso,
em cada censo, com tabulações cruzadas por sexo, raça, escolaridade e região. Esse
nível revela as diferenças brutas, mas não distingue se decorrem da religião ou da
composição demográfica de cada grupo.

**(b) Coortes com controles.** As comparações passam a ser feitas dentro de grupos
demograficamente homogêneos: mesma janela de nascimento, sexo, raça, escolaridade,
isolando progressivamente a variável religiosa.

**(c) Modelos de regressão.** Dois modelos formalizam o teste das hipóteses:

- **OLS (mínimos quadrados ordinários)** sobre o logaritmo da renda em salários
  mínimos: `log(renda) ~ religião + idade + idade² + sexo + raça + diploma + região + ano`.
  O uso do logaritmo permite ler os coeficientes como diferenças percentuais
  aproximadas de renda entre grupos, mantidas constantes as demais características.
  A categoria de referência é a católica, maior grupo do país.

- **Regressão logística** sobre a probabilidade de diploma superior na geração de
  "religião de berço": `P(diploma) ~ religião + idade + sexo + raça + região`.
  Os resultados são reportados em razões de chance (odds ratios), em que valores
  acima de 1 indicam chance maior de diploma que a referência católica, e abaixo
  de 1, chance menor.

A robustez da associação religião × escolaridade é verificada adicionalmente por
teste qui-quadrado de independência. Adota-se significância de 5%, com a maioria
dos resultados centrais apresentando p < 0,001.

### 2.3 Implementação em Python

Todo o trabalho foi implementado em Python, em quatro notebooks Jupyter que formam
um pipeline reprodutível da coleta dos dados brutos à modelagem estatística. O código
completo está disponível em
[github.com/b-scobino/renda-e-religiao](https://github.com/b-scobino/renda-e-religiao).
Principais bibliotecas: `pandas` (manipulação de dados), `numpy` (operações numéricas),
`requests` (download), `dbfread` (leitura de arquivos dBase), `plotly` e `ipywidgets`
(visualização interativa), `scipy` e `statsmodels` (estatística e regressões).

**Notebook 01 — Coleta.** Automatiza o download dos microdados a partir do FTP público
do IBGE. As URLs dos 58 arquivos (1 nacional de 1991; 28 de 2000; 29 de 2010, com São
Paulo dividido em dois) são montadas dinamicamente com *f-strings* e *list
comprehensions* a partir da lista de siglas das UFs. A função `baixar_arquivo()` usa
`requests` com `stream=True`: o arquivo é gravado em blocos de 8 KB, com barra de
progresso via `tqdm`, evitando carregar gigabytes na memória, e verificando com
`os.path.isfile()` se o arquivo já existe, tornando a execução retomável. A
descompactação usa o utilitário 7-Zip via `subprocess`, necessário porque os arquivos
do IBGE empregam um método de compressão antigo não suportado pelo módulo `zipfile`
nativo.

**Notebook 02 — Limpeza e harmonização.** O maior desafio técnico do trabalho: cada
censo usa formato e codificação distintos.

- *1991*: arquivos dBase (`.DBF`), lidos com `dbfread`; com milhões de registros por
  arquivo, a leitura integral é inviável, e `itertools.islice` limita a amostra a
  5.000 registros por UF (135 mil no total do censo).
- *2000 e 2010*: arquivos texto de largura fixa, sem separadores. As posições de cada
  variável foram extraídas da documentação oficial: script SAS de 2000; planilha de
  layout de 2010. O fatiamento é feito por índice de string (`linha[pos-1:pos-1+tam]`).
- *Classificação religiosa*: os códigos de religião (2 dígitos em 1991; 3 dígitos em
  2000/2010, com estrutura hierárquica em que o prefixo indica o grupo) foram mapeados
  para 9 grupos comparáveis entre censos por funções condicionais
  (`classificar_religiao_1991` e `classificar_religiao_3dig`), construídas a partir dos
  dicionários oficiais do IBGE. A validação usou proporções históricas conhecidas
  e validadas por pesquisa web (ex.: espíritas ≈ 1,1% em 1991; pentecostais ≈ 13,3%
  em 2010).
- *Harmonização*: renda convertida a salários mínimos da época (em 2010, o campo traz
  5 casas decimais implícitas, exigindo divisão por 100.000); escolaridade unificada
  na variável binária "superior completo". Os três censos são empilhados com
  `pd.concat()` e gravados em um único CSV de 410 mil registros.

**Notebook 03 — Análise.** Estatísticas descritivas e coortes com `groupby()`/
`median()`/`mean()`, filtragem booleana com máscaras do pandas e a função reutilizável
`filtrar_coorte()`. Os gráficos combinam `plotly.express` (linhas e barras com *hover*
unificado) e `ipywidgets.interact`: *dropdowns* de gênero, escolaridade, região e
etnia que redesenham o gráfico a cada seleção. Recortes com menos de 10 observações
são suprimidos automaticamente dos gráficos (condição `n >= 10` no código): medianas
calculadas sobre pouquíssimos indivíduos são dominadas por casos atípicos e
transmitiriam ao leitor, com a mesma autoridade visual de grupos com milhares de
observações, padrões que podem ser mero acaso amostral.

**Notebook 04 — Regressões.** Os modelos usam a interface de fórmulas do `statsmodels`
(`smf.ols` e `smf.logit`), na qual a especificação é declarada como texto.
Exemplificando, `log_renda ~ C(grupo_religioso) + idade + I(idade**2) + ...`. O
marcador `C()` declara variáveis categóricas: o pacote as converte internamente em
indicadoras binárias, as variáveis *dummies*, comparando cada grupo religioso à
categoria de referência (católica, maior grupo religioso do Brasil). O termo
`I(idade**2)` adiciona a idade ao quadrado, capturando o formato não linear do ciclo
de renda, que cresce até a meia-idade e declina em seguida. Na regressão logística,
os coeficientes (em escala de log-chances) são exponenciados com `np.exp()` para
obtenção das razões de chance (*odds ratios*), diretamente interpretáveis. Por fim,
a robustez da associação religião × diploma é verificada por teste qui-quadrado
(`scipy.stats.chi2_contingency`) aplicado à tabela de contingência gerada por
`pd.crosstab()`: o teste compara as contagens observadas com as que seriam esperadas
caso religião e escolaridade fossem independentes.

### 2.4 Limitações

Quatro limitações merecem registro:

(a) Os censos não permitem acompanhar indivíduos; as coortes sintéticas seguem grupos,
e mudanças de composição religiosa interna (conversões) misturam-se às trajetórias de
renda.

(b) O dicionário oficial do layout DBF de 1991 não documenta a codificação de religião
encontrada nos dados; o mapeamento foi inferido pela estrutura de faixas e validado
contra proporções históricas, com possível imprecisão em grupos pequenos.

(c) A amostragem fixa por UF (5.000 registros) superrepresenta estados pequenos,
motivo pelo qual todas as análises controlam por região.

(d) Associação não é causalidade: a renda dos pais é candidata óbvia a explicar tanto
a religião de berço quanto o destino educacional.

```python
# Preparo da base para as tabelas e figuras desta seção

import os
import pandas as pd
import numpy as np
import plotly.express as px

caminho_dados = "https://raw.githubusercontent.com/b-scobino/renda-e-religiao/main/censo_religiao_renda.csv"
df = pd.read_csv(caminho_dados)

def definir_regiao(uf):
    primeiro = str(uf)[0]
    return {"1": "Norte", "2": "Nordeste", "3": "Sudeste",
            "4": "Sul", "5": "Centro-Oeste"}.get(primeiro, "Indefinida")

df["regiao"] = df["uf"].apply(definir_regiao)
df["sexo_label"] = df["sexo"].map({1: "Masculino", 2: "Feminino"})
df["raca_label"] = df["raca"].map({1: "Branca", 2: "Preta", 3: "Amarela", 4: "Parda", 5: "Indígena"})

def tem_superior(linha):
    if linha["ano"] in [1991, 2000]:
        return 15 <= linha["escolaridade_original"] <= 17
    return linha["escolaridade_original"] == 4

grupos_foco = ["Católica", "Evangélica Tradicional", "Evangélica Pentecostal",
               "Sem religião", "Mediúnica/Espírita"]

df_renda = df[(df["idade"] >= 15) & (df["idade"] <= 70) &
              (df["renda_sm"] > 0) & (df["renda_sm"] <= 200) &
              (df["grupo_religioso"].isin(grupos_foco))].copy()
df_renda["superior"] = df_renda.apply(tem_superior, axis=1)

print(f"Base analítica: {len(df_renda):,} registros com renda positiva".replace(",", "."))
```

## 3. Resultados

### 3.1 A transição religiosa (1991–2010)

A amostra reproduz com fidelidade a transformação religiosa do período. Entre as
crianças de 5 a 15 anos em 1991, a "religião de berço" da geração 1976–86, 91%
eram católicas e apenas 2,6% pentecostais. Na mesma geração, adulta em 2010, os
católicos haviam caído para 71,7% e os pentecostais saltado para 13,8%:
quintuplicaram em uma geração. É sobre essa população em movimento que as análises
seguintes operam.

### 3.2 Renda por religião: o retrato bruto

Antes de qualquer controle, as diferenças de renda entre grupos religiosos são
expressivas e estáveis nos três censos:

```python
# Tabela 3.1 — Renda por grupo religioso, por censo (sem controles)

for ano in [1991, 2000, 2010]:
    dados = df_renda[df_renda["ano"] == ano]
    n = dados.groupby("grupo_religioso")["renda_sm"].count()
    media = dados.groupby("grupo_religioso")["renda_sm"].mean().round(2)
    mediana = dados.groupby("grupo_religioso")["renda_sm"].median().round(2)
    tabela = pd.DataFrame({"n": n, "média (SM)": media, "mediana (SM)": mediana})
    print(f"\n===== {ano} =====")
    print(tabela.sort_values("mediana (SM)", ascending=False))
```

O padrão bruto é consistente: espíritas no topo (mediana 3 a 4 vezes a católica),
pentecostais na base, sem religião e tradicionais acima da média. Lido isoladamente,
o quadro sugeriria que a fé pentecostal "empobrece" e a ausência de fé cristã ou ausência
total de fé "enriquece".

As seções seguintes mostram que essa leitura é precipitada: os grupos religiosos têm
composições demográficas muito distintas, e é a composição, não a fé, que explica a
maior parte do que se vê acima.

### 3.3 Coortes: a mesma diferença, dois mundos

Acompanhando a geração nascida entre 1956 e 1966 nos três censos, o efeito da
composição torna-se visível. No topo da pirâmide (homens brancos com ensino superior),
"sem religião" lidera a renda com folga nos três censos, e em 1991 simplesmente não
havia pentecostais nesse perfil na amostra. Na base (mulheres negras sem ensino
superior), o quadro muda: evangélicas tradicionais lideram, e a diferença entre
pentecostais e católicas praticamente desaparece. Entre homens da base, pentecostais
chegam a superar católicos e sem religião.

A relação entre religião e renda, portanto, não é uniforme: depende de classe e de
gênero. A "desvantagem pentecostal" do retrato bruto concentra-se nos estratos altos.
Na base, onde o pentecostalismo de fato recruta, ela é pequena ou inexistente.

```python
# Tabela 3.2 — Geração de berço (1976-86) em 2010: renda por religião e escolaridade

adultos_2010 = df_renda[(df_renda["ano"] == 2010) &
                        (df_renda["idade"] >= 24) & (df_renda["idade"] <= 34)].copy()

print("===== COM ensino superior =====")
com_s = adultos_2010[adultos_2010["superior"] == True]
print(pd.DataFrame({
    "n": com_s.groupby("grupo_religioso")["renda_sm"].count(),
    "mediana (SM)": com_s.groupby("grupo_religioso")["renda_sm"].median().round(2)
}).sort_values("mediana (SM)", ascending=False))

print("\n===== SEM ensino superior =====")
sem_s = adultos_2010[adultos_2010["superior"] == False]
print(pd.DataFrame({
    "n": sem_s.groupby("grupo_religioso")["renda_sm"].count(),
    "mediana (SM)": sem_s.groupby("grupo_religioso")["renda_sm"].median().round(2)
}).sort_values("mediana (SM)", ascending=False))
```

### 3.4 O leque do diploma

A tabela acima contém o achado descritivo central do trabalho. **Sem o diploma, a
religião quase não diferencia a renda**, as medianas dos grupos são praticamente
idênticas, em torno de 1 salário mínimo. **Com o diploma, o leque se abre**: a mesma
credencial educacional está associada a retornos muito distintos conforme o grupo
religioso, com "sem religião" e espíritas extraindo do diploma um múltiplo de renda
consideravelmente maior que o dos pentecostais.

E o acesso ao próprio diploma é fortemente desigual entre religiões de berço: na
geração 1976–86, criada majoritariamente antes da transição religiosa, 18,6% dos que
cresceram em lares evangélicos tradicionais concluíram o ensino superior, contra 12,8%
dos sem religião, 11% dos católicos e apenas 4,6% dos pentecostais, razão de 4
para 1 entre os dois ramos do protestantismo.

No extremo superior, os espíritas constituem caso à parte: mais da metade do grupo
(51,2%) chegou ao diploma, a maior taxa entre todas as afiliações.

O padrão central sobrevive ao controle de raça e região: entre pardos, a taxa
tradicional (15,3%) triplica a católica (5,1%). Já o grupo "sem religião" revela-se heterogêneo sob o controle racial: lidera entre brancos (23,9% de diplomados), mas cai à penúltima posição entre pretos (3,3%), sugerindo que a categoria abriga dois perfis sociais distintos: o secularismo
escolarizado das camadas altas e o desligamento institucional nas periferias.

### 3.5 Regressões: o que sobrevive aos controles

Os modelos formais quantificam o que as coortes sugerem. Na regressão OLS sobre o
logaritmo da renda, controlando idade, sexo, raça, diploma, região e ano, os
coeficientes de religião (referência: católica; todos com p < 0,001) são:

| Grupo religioso | Coeficiente | Efeito aproximado sobre a renda |
|---|---|---|
| Mediúnica/Espírita | +0,502 | **+65,2%** |
| Sem religião | +0,100 | +10,5% |
| Evangélica Tradicional | +0,083 | +8,7% |
| Evangélica Pentecostal | −0,032 | **−3,2%** |

A Hipótese 1 confirma-se: pentecostais ganham menos, mesmo entre demograficamente
idênticos. Mas o efeito é modesto: **3,2%**. No dado bruto, a desvantagem mediana
pentecostal superava 30%; com controles, quase toda ela se dissolve. É a Hipótese 2
em ação: a maior parte da "pobreza pentecostal" é composição do grupo (seleção),
não efeito da fé. Para escala: o diploma associa-se a +375% de renda; ser mulher,
a −35%; residir no Nordeste, a −56%. **Diante de educação, gênero e geografia, a
religião é coadjuvante.**

O canal educacional, porém, conta outra história. Na regressão logística sobre a
probabilidade de diploma na geração de berço (referência: católica), as razões de
chance são:

| Religião de berço | Odds ratio | Leitura |
|---|---|---|
| Mediúnica/Espírita | 4,78 | quase 5× a chance de diploma |
| Evangélica Tradicional | 1,70 | +70% de chance |
| Sem religião | 1,28 | +28% de chance |
| Evangélica Pentecostal | **0,45** | **menos da metade da chance** |

Todos os efeitos com p < 0,01, controlando idade, sexo, raça e região. O teste
qui-quadrado corrobora (χ² = 740,7; p < 10⁻¹⁵⁸). Crescer em lar pentecostal
associa-se a menos da metade da chance de diploma de um católico do mesmo perfil
demográfico. E, dado que o diploma é o maior multiplicador de renda do modelo,
é por essa porta que a religião de berço efetivamente pesa sobre o destino econômico.

## 4. Conclusão

Pertencer a uma religião pode ser benéfico para a renda?

Os dados dos três censos respondem com uma distinção que o debate público raramente faz:
**a fé professada altera pouco o salário; a fé em que se nasce pesa muito sobre o
caminho até ele.**

A Hipótese 1 confirmou-se, mas com efeito modesto: controlando idade, sexo, raça,
escolaridade, região e ano, pentecostais ganham cerca de 3% menos que católicos
demograficamente idênticos, ainda a maioria do país. Uma fração da desvantagem superior a 30% observada no dado bruto.

A dissolução da diferença sob controles é precisamente o que previa a Hipótese 2:
a "pobreza pentecostal" é majoritariamente composição do grupo. O pentecostalismo
brasileiro recrutou, e segue recrutando, a base da pirâmide social.

Seus fiéis não empobreceram ao entrar, entraram porque a igreja cresceu onde a
pobreza já estava. Pelo mesmo raciocínio, a renda superior dos sem religião e dos
espíritas reflete, em grande medida, quem são: mais brancos, mais urbanos, mais
escolarizados. Não se trata de um prêmio da descrença ou da mediunidade.

Quanto à Teologia da Prosperidade, os dados não lhe são generosos: não há prêmio de
renda associado à adesão pentecostal em nenhum recorte; no estrato mais alto, há
penalidade. A fé, por forte que seja, não aparece nos contracheques ou rendimentos.

A hipótese secundária sobre a análise weberiana, contudo, encontrou eco surpreendente.
Protestantes tradicionais apresentam vantagem educacional robusta que atravessa raça
e região, triplicando a taxa católica de diplomas entre pessoas pardas, por exemplo. Além disso,
no Nordeste o resultado se reafirma, padrão compatível com o papel histórico de
denominações que fundaram escolas e valorizaram a alfabetização como acesso direto à
Escritura Sagrada.

O contraste com o ramo pentecostal (razão de 4 para 1 nos diplomas; odds ratio de
0,45 contra 1,70) sugere que a pergunta de Weber continua viva, mas exige a precisão
que ele próprio praticava: não "a fé enriquece?", e sim qual fé, por qual mecanismo
e para quem.

O mecanismo, os dados indicam com clareza, é educacional. Crescer em lar pentecostal
associa-se a menos da metade da chance de concluir o ensino superior; crescer
espírita, a quase cinco vezes mais que o maior grupo de brasileiros (católicos). E o
diploma multiplica a renda por fator próximo de cinco. A religião de berço opera como
condição de partida: não define o salário, mas pesa na porta de entrada do principal
elevador de mobilidade social, o estudo formal.

Permanecem questões que este trabalho não responde. A renda dos pais, ausente dos
microdados do censo, é candidata natural a explicar parte da associação entre
religião de berço e diploma.

A heterogeneidade dos "sem religião", que lideram entre brancos e despencam entre
pretos, pede investigação própria. O prêmio espírita de +65%, que sobrevive a todos
os controles, sugere mecanismos específicos: profissionais, de rede, de capital
cultural, que mereceriam dados individuais. E a chegada dos microdados do Censo 2022
permitirá testar se a geração criada já dentro da transição religiosa reproduz os
padrões aqui encontrados.

Deus, deuses ou espíritos de luz, ao que tudo indica, não ajudam nem atrapalham a ter
um bom salário. Mas a fé em que se cresce, pelo que os dados mostram, abre ou estreita o
caminho até ele.
